# Why Quantum Machine Learning?

The application of quantum computing to the computation of machine learning algorithms is known as quantum machine learning.

Recently, we have seen algorithms defeat global chess and go champions and learn to drive vehicles. Almost every industry can be found using machine learning, from the military to aerospace, from manufacturing to agriculture, and from finance to healthcare.

But because these algorithms have billions of parameters, training them gets harder and harder. Such problems are unsolvable with present computing technology, but quantum computers promise to solve them. They can also carry out an infinite number of superposed tasks concurrently thanks to their ability to calculate many states at once. Thus, quantum computing has the potential to accelerate and enhance machine learning methods.

Quantum computing, as opposed to traditional computers based on sequential information processing, makes advantage of the superposition, entanglement, and interference phenomena in quantum physics. But it decreases the amount of processing power required to solve a problem rather than increasing the amount of computing power accessible.

But, quantum computing calls for a shift in how we view computers. A completely new set of algorithms are necessary, quantum information-encoding and quantum processingÂ using algorithms. This includes machine learning algorithms.

Quantum machine learning has the potential to be revolutionary. Although the conceptual aspects of this combination of machine learning and quantum computing, two areas of active study, are still substantially unexplored, there are already several examples where it is being used to address practical issues. A large number of high-tech startups, including Google, Amazon, IBM, Microsoft, and others, compete to develop and market the first quantum machine learning systems.

## Working of Machine Learning Algorithm

Each machine learning algorithm consists of three parts:

1. The representation shows the internal structure of the algorithm's knowledge representation. It could include guidelines, examples, decision trees, SVMs, neural networks, and more.

2. The function of assessment is to assess potential algorithm parameterizations. A few examples are cost, margin, entropy, posterior probability, squared error, accuracy, prediction and recall, and others.

3. The optimization explains how to produce candidate parameterizations for algorithms. It is referred to as the search process and includes techniques like constrained optimization, convex optimization, and combinatorial optimization.

## Real Power of Quantum Computers

The realm of quantum mechanics is distinct from the physics we encounter on a daily basis. In this way, traditional (digital) computing and the world of quantum computing are dissimilar.

Its processing speed is not what makes quantum computing so strong. It processes fairly slowly. Its memory is also not what makes quantum computing so potent. It's absurdly small. A few quantum bits are what we're referring to.

The algorithms that quantum computing makes possible display different complexity characteristics from their classical counterparts, which is what makes them so strong.

Superposition can be used by a good quantum algorithm (like Shor's algorithm) to examine all potential factors of a number at once. Moreover, interference is used to combine all viable solutions in a way that results in the right answer rather than simply computing the outcome.

Because it promises to do specific kinds of mathematical computations with less complexity, quantum computing is extremely potent.

Considering what current traditional machine learning algorithms are capable of. If we anticipate quantum computing to significantly lower the difficulty of training these algorithms, then the hype about quantum machine learning becomes more reasonable.

### Coping with Missing Values

```python
train_data = train_data.dropna(subset=["Embarked"])
# Only two of our passengers do not have it. It's tolerable

train_data = train_data.drop("Cabin", axis=1)
# As there aren't many details available concerning the cabin, let's disregard it.

mean = train_data["Age"].mean()
train_data["Age"] = train_data["Age"].fillna(mean)
# Several times, the age is missing. 
# Nonetheless, intuition suggests that it might be crucial for someone's likelihood of survival.

train_data.info()
```

The column ("Embarked") is a named parameter subset for the dropna method.
The columns that determine whether to eliminate the row are specified by this option (passenger). The row is eliminated if at least one value in each of these columns is missing.

With the technique drop, we delete the entire column. The column (Cabin) that we want to remove is then provided as a positioned parameter. The named option axis value 1 we provide indicates that we want to remove the entire column.

Using the fillna technique, we fill in the missing information. We call this function on the Age column rather than the entire DataFrame since we just want to fill in the missing values in this column. The value we want to set is provided as an argument. This is the average age of all passengers that we previously calculated.

We now have ten columns, 889 rows, and no more missing data.

### Identifying unique values in columns

The greatest thing we could do is memorise all of the passengers and whether they survived if our only objective was to anticipate labels we already know. Yet memorization must be avoided if we're going to design an algorithm that works well even with unknowable input.


We haven't even begun to develop our algorithm yet. Nevertheless, because our data contains potential identifiers, the features we utilise in our algorithm have an impact on whether the system can memorise data.

Let's look at how many distinct values are present in each of these columns:

```python
print('There are {} different (unique) PassengerIds in the data'.format(train_data["PassengerId"].nunique()))

print('There are {} different (unique) names in the data'.format(train_data["Name"].nunique()))

print('There are {} different (unique) ticket numbers in the data'.format(train_data["Ticket"].nunique()))
```

The two ideal IDs are Name and PassengerId. Hence, there is a distinct value for each of the 889 rows in our dataset.


Moreover, there are 680 unique ticket numbers. Family tickets can be a reason why the Ticket isn't a great identifier. Nonetheless, a conclusion drawn from this data seems to favour memorising over gaining applicable ideas


We remove these columns.

```python
train_data = train_data.drop("PassengerId", axis=1)

train_data = train_data.drop("Name", axis=1)

train_data = train_data.drop("Ticket", axis=1)

train_data.info()
```

### Transforming textual data into numbers

Numerous machine learning methods only use numbers in their operations. Textual information must be converted into numerical form if it is to be used.


For this task, Scikit-Learn offers a transformer with the name of LabelEncoder.

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()

for col in ['Sex', 'Embarked']:

    LE.fit(train_data[col])

    train_data[col] = LE.transform(train_data[col])

train_data.head()
```

### Feature Scaling

Algorithms for machine learning typically operate on numbers with uniform scales. If distinct numbers have different scales, the algorithm might give more weight to the numbers with higher scales.

The maximum value in a column is returned by the max method. As we can see, the highest fare was around 512 and the oldest passenger was 80.

```python
print('The maximum age is {}'.format(train_data["Age"].max()))

print('The maximum fare is {}'.format(train_data["Fare"].max()))
```

Min-Max-Scaling, commonly referred to as normalisation, is a typical technique for dealing with data on various scales. Values are shifted and rescaled in this procedure so that they end up falling between 0 and 1. Each value is subtracted from the minimum value and divided by the maximum value minus the minimum value.

To help us with this, Scikit-Learn offers the MinMaxScaler transformer.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(train_data)

train_data = scaler.transform(train_data)

print('The minimum value is {} and the maximum value is {}'.format(train_data.min(), train_data.max()))
```

### Training and Testing

The objective of creating an algorithm that excels at using data it already knows and anticipates the labels of data it hasn't seen yet was already described. Because of this, it's crucial to divide the data into a training set and a testing set. We create our algorithm using the training set. And we test the performance of it on the testing set.

We would prefer save a few rows from the Kaggle training set for testing in order to keep things straightforward and perform the validation ourselves.


A test set can be easily divided. Scikit-learn offers a helpful approach for it as well. It's called train_test_split.

```python
from sklearn.model_selection import train_test_split

input_data = train_data[:, 1:8]

labels = train_data[:, 0]

train_input, test_input, train_labels, test_labels = train_test_split(input_data, labels, test_size = 0.2)

print('We have {} training and {} testing rows'.format(train_input.shape[0], test_input.shape[0]))

print('There are {} input columns'.format(train_input.shape[1]))
```

The data is now saved to the filesystem.

```python
import numpy as np

with open('/Volumes/MY_PASSPORT/IIT_Delhi_QCML/train.npy', 'wb') as f:
    np.save(f, train_input)
    np.save(f, train_labels)

with open('/Volumes/MY_PASSPORT/IIT_Delhi_QCML/test.npy', 'wb') as f:
    np.save(f, test_input)
    np.save(f, test_labels)
```
